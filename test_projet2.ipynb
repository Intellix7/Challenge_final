```python
import numpy as np
import matplotlib.pyplot as plt

# Huggingface datasets
from datasets import load_dataset, get_dataset_split_names, Image

recycling_dataset = load_dataset('viola77data/recycling-dataset', split='train').cast_column("image", Image(decode=False))
```

```python
labels = recycling_dataset.features["label"].names
```

```python
recycling_dataset[0]["image"]["path"]
```

```python
import os
import cv2
import random
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions

model = ResNet50(weights='imagenet')

n_img = 2

# img_dic = 'images/'
#
# img_paths = os.listdir(img_dic)
# print(img_paths)

mean = 0

for row in recycling_dataset.random.shuffle()[:n_img]:
    img = cv2.imread(row["image"]["path"])
    img = cv2.resize(img, (224, 224))  # Resize the image to match the model's input size
    x = image.img_to_array(img)  # Convert the image to a numpy array
    x = np.expand_dims(x, axis=0)  # Add a batch dimension
    x = preprocess_input(x)

    # Make predictions
    preds = model.predict(x)
    print(preds)

    # Decode and display predictions
    print('Predicted:', decode_predictions(preds, top=3)[0], 'label = ', row)

    mean += preds[0][0]

print(f"Average confidence score : {mean / n_img}")
```
